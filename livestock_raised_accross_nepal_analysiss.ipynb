* Merge them into single data source

```python


# List of all 8 CSV files
files = [
    'wool-production-in-nepal-by-district.csv',
    'yak-nak-chauri-population-in-nepal-by-district.csv',
    'horseasses-population-in-nepal-by-district.csv',
    'net-meat-production-in-nepal-by-district.csv',
    'milk-animals-and-milk-production-in-nepal-by-district.csv',
    'production-of-egg-in-nepal-by-district.csv',
    'production-of-cotton-in-nepal-by-district.csv',
    'rabbit-population-in-nepal-by-district.csv'
]

# Read each file into a list of DataFrames
raw_dfs = [pd.read_csv(file) for file in files]
raw_dfs[0].head()
```

```python




name_mapping = {
    'SANKHUWASHAVA': 'SANKHUWASABHA',
    'TERATHUM': 'TERHATHUM',
    'RAMECHAP': 'RAMECHHAP',
    'Dang': 'DANG',
    'Banke': 'BANKE',
    'Bardiya': 'BARDIYA'
}

# Regex to detect and remove aggregate (non-district) rows
aggregate_pattern = re.compile(r'\.|REGION|MOUNTAIN|HILLS|TERAI|NEPAL|Total', re.IGNORECASE)

# Process each DataFrame separately for name standardization
standardized_dfs = []
for df in raw_dfs:
    # Convert to uppercase and strip whitespace
    df['DISTRICT'] = df['DISTRICT'].str.strip().str.upper()
    
    # Apply spelling corrections
    df['DISTRICT'] = df['DISTRICT'].replace(name_mapping)
    
    standardized_dfs.append(df)

    print(df.head())
```

```python
clean_dfs = []

for df in standardized_dfs:
    # Remove aggregate rows (regional summaries, totals)
    df = df[~df['DISTRICT'].str.contains(aggregate_pattern, na=False)]
    
    # Set DISTRICT as index and remove any duplicate districts
    df = df.set_index('DISTRICT')
    df = df[~df.index.duplicated(keep='first')]
    
    clean_dfs.append(df)

# show the cleaned    tables
for df in clean_dfs:
    print(df.head())    
```

```python
# Merge all cleaned DataFrames using outer join to keep all 75 districts
merged_df = pd.concat(clean_dfs, axis=1, join='outer')

# Convert all value columns to numeric (handle any non-numeric entries)
for col in merged_df.columns:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

# Fill missing values with 0 (unreported livestock/production = 0)
merged_df = merged_df.fillna(0)

# Sort districts alphabetically for better readability
merged_df = merged_df.sort_index()

# Final result
print(f"Final shape: {merged_df.shape}") 
merged_df.iloc[0:40]
```

```python
# 1. Read all data sources

horse = pd.read_csv('horseasses-population-in-nepal-by-district.csv')
milk = pd.read_csv('milk-animals-and-milk-production-in-nepal-by-district.csv')
meat = pd.read_csv('net-meat-production-in-nepal-by-district.csv')
cotton = pd.read_csv('production-of-cotton-in-nepal-by-district.csv')
egg = pd.read_csv('production-of-egg-in-nepal-by-district.csv')
rabbit = pd.read_csv('rabbit-population-in-nepal-by-district.csv')
wool = pd.read_csv('wool-production-in-nepal-by-district.csv')
yak = pd.read_csv('yak-nak-chauri-population-in-nepal-by-district.csv')

datasets = [horse, milk, meat, cotton, egg, rabbit, wool, yak]


# 2. Standardize column names

for df in datasets:
    df.columns = (
        df.columns
        .str.strip()
        .str.upper()
        .str.replace(' ', '_')
    )


# 3. Clean and standardize district names

def clean_district_names(df):
    df['DISTRICT'] = (
        df['DISTRICT']
        .str.strip()
        .str.upper()
    )
    
    # Fix known spelling inconsistencies
    corrections = {
         'SANKHUWASHAVA': 'SANKHUWASABHA',
    'TERATHUM': 'TERHATHUM',
    'RAMECHAP': 'RAMECHHAP',
    'Dang': 'DANG',
    'Banke': 'BANKE',
    'Bardiya': 'BARDIYA'
    }
    df['DISTRICT'] = df['DISTRICT'].replace(corrections)
    
    return df

datasets = [clean_district_names(df) for df in datasets]


# 4. Remove aggregate / regional rows

aggregate_keywords = [
    'REGION', 'NEPAL', 'TOTAL',
    'E.MOUNTAIN', 'E.HILLS', 'E.TERAI',
    'C.MOUNTAIN', 'C.HILLS', 'C.TERAI',
    'W.MOUNTAIN', 'W.HILLS', 'W.TERAI'
]

def remove_aggregate_rows(df):
    mask = df['DISTRICT'].str.contains('|'.join(aggregate_keywords))
    return df[~mask]

datasets = [remove_aggregate_rows(df) for df in datasets]


# 5. Merge all datasets on DISTRICT

merged_df = reduce(
    lambda left, right: pd.merge(left, right, on='DISTRICT', how='outer'),
    datasets
)


# 6. Handle missing values

# Missing values indicate no reported population/production
merged_df.fillna(0, inplace=True)


# 7. Final dataset overview

print("Final shape of merged dataset:", merged_df.shape)
print("Number of unique districts:", merged_df['DISTRICT'].nunique())
merged_df.iloc[:75]
```

When it comes to meat production, Chitwan emerges as the undisputed champion of Nepal. Known for its commercial poultry and livestock farms, Chitwan contributes significantly to the national protein supply.
The Pattern: Nationally, Chevon (Goat meat) and Chicken dominate the market. Chevon represents a massive portion of the meat distribution, highlighting the cultural and economic importance of goat farming in the Nepalese hills and plains.

Milk production tells a different story. While Chitwan is the meat capital, Kavre leads the nation in milk production.
Efficiency Insights: Interestingly, high volume doesn't always mean high efficiency. When we look at "Milk per Cow," districts like Syangja and Kanchanpur lead the way, suggesting better breeds or more intensive management practices in these areas.
Buffalo vs. Cow: In most districts, Buffalo milk still outweighs Cow milk, reinforcing the buffalo's status as the "Black Gold" of rural Nepal.

As we move to the mountainous regions, the livestock profile shifts. Districts like Jumla and Mugu are the "Wool Hubs" of Nepal.
The Pattern: Wool production is almost exclusively concentrated in the high-altitude districts of the Mid-Western and Western regions, where sheep and Yak/Chauri farming are not just industries, but a way of life.
